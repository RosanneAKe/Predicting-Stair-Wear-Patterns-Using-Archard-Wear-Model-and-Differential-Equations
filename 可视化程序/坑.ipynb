```python
import numpy as np
import matplotlib.pyplot as plt
# 1. 
L = 1.2  # 
D = 0.3  # 
grid_size = 100  # 

# 2. 
x = np.linspace(0, L, grid_size)  #  
y = np.linspace(0, D, grid_size)  #  
X, Y = np.meshgrid(x, y)  # 
def wear_depth(x, y, x0=0.6, y0=0.3, r=0.15):
    #  (x, y) 
    distance_to_arc = np.sqrt((x - x0)**2 + (y - y0)**2) - r
    
    # 
    depth = -1 * np.exp(-distance_to_arc**2 / 0.05**2)  # 
    return depth
# 4. 
W = wear_depth(X, Y)  # 

# 5. 2D
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, W, levels=20, cmap='viridis')
plt.colorbar(label="Wear Depth (cm)")
plt.xlabel("Width (m)")
plt.ylabel("Depth (m)")
plt.title("Tread Surface Wear Depth")
plt.show()

# 6. 3D
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, W, cmap='viridis', edgecolor='none')
ax.set_xlabel("Width (m)")
ax.set_ylabel("Depth (m)")
ax.set_zlabel("Wear Depth (cm)")
ax.set_title("3D View of Tread Wear Depth")
plt.show()

#
print("Wear depth data shape:", W.shape)
```
