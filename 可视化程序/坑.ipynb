```python
import numpy as np
import matplotlib.pyplot as plt
# 1. 定义台阶踏面
L = 1.2  # 台阶宽度
D = 0.3  # 台阶进深
grid_size = 100  # 网格划分

# 2. 生成网格
x = np.linspace(0, L, grid_size)  # 横坐标 
y = np.linspace(0, D, grid_size)  # 纵坐标 
X, Y = np.meshgrid(x, y)  # 二维网格
def wear_depth(x, y, x0=0.6, y0=0.3, r=0.15):
    # 计算点 (x, y) 到圆弧的水平距离
    distance_to_arc = np.sqrt((x - x0)**2 + (y - y0)**2) - r
    
    # 磨损深度函数：离圆弧越远，凹陷越浅
    depth = -1 * np.exp(-distance_to_arc**2 / 0.05**2)  # 可根据需要调整参数
    return depth
# 4. 计算磨损深度
W = wear_depth(X, Y)  # 磨损深度

# 5. 可视化2D热图
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, W, levels=20, cmap='viridis')
plt.colorbar(label="Wear Depth (cm)")
plt.xlabel("Width (m)")
plt.ylabel("Depth (m)")
plt.title("Tread Surface Wear Depth")
plt.show()

# 6. 可视化3D图
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, W, cmap='viridis', edgecolor='none')
ax.set_xlabel("Width (m)")
ax.set_ylabel("Depth (m)")
ax.set_zlabel("Wear Depth (cm)")
ax.set_title("3D View of Tread Wear Depth")
plt.show()

#输出结果
print("Wear depth data shape:", W.shape)
```
