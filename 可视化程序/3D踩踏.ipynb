```python
# Adjusting the model to include four A parameters for individual control

# Parameters
mu1, mu2, mu3, mu4 = 50, 70, 20, 15  # Updated mu3 and added mu4 for symmetry
sigma1, sigma2, sigma3, sigma4 = 8, 6, 4, 800
A1, A2, A3, A4 = 0.8, 1.2, 1, 1  # Four independent amplitudes
lambda_ = 2  # Exponential decay rate for the front edge wear

# Create grid
x = np.linspace(0, 100, 200)
y = np.linspace(0, 30, 200)
X, Y = np.meshgrid(x, y)

# z-x function with double-peak depression
Z_x = -A1 * np.exp(-(X - mu1)**2 / (2 * sigma1**2)) - A2 * np.exp(-(X - mu2)**2 / (2 * sigma2**2))

# Front edge wear (Exponential decay)
front_wear = -A3 * np.exp(-lambda_ * Y)

# Inner wear (Gaussian distribution)
inner_wear = -A4 * np.exp(-(Y - mu3)**2 / (2 * sigma3**2))

# Total wear combining z-x and z-y functions
Z = 19 + Z_x + front_wear + inner_wear

# Plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set axis limits to enlarge the scale
ax.set_xlim([0, 100])  # Increase x-axis range
ax.set_ylim([0, 30])   # Increase y-axis range
ax.set_zlim([0, 20])    # Set z-axis to [0, 20] to ensure it includes the 19 plane

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Tread Surface Wear (Double Peaks + Front Edge & Inner Wear)")
plt.show()
```
